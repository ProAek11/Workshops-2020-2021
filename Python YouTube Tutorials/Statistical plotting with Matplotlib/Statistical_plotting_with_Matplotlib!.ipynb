```
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12,10)
plt.rcParams.update({'font.size': 14})

from scipy import stats
import numpy as np
import pandas as pd
```

# **Uniform Distribution**

```
plt.hist(stats.uniform.rvs(size = 10000), edgecolor = 'black', color = 'lightgreen')
plt.title('Uniform Distribution $U \sim (0,1)$')
plt.ylabel('Number of Observations');
```

# **Poisson Distribution**

```
fig, axs = plt.subplots(4)
fig.suptitle('Poisson Distributions')
fig.tight_layout(rect=[0, 0, 1.05, 0.95])
color_list = ['r', 'g', 'b', 'purple']
bins = np.arange(11) - .5

for i,j in zip(range(4), color_list):
  axs[i].hist(stats.poisson.rvs(mu = i +1, size = 100), bins = bins, color = j, edgecolor = 'black', label = f'$\lambda = {i + 1}$')

fig.legend(bbox_to_anchor=(1, .92));
```

# **Normal Distribution**

```
mean = 0
variance = 1
st_dev = variance**.5
norm_samples = np.random.normal(size = 10000)

plt.hist(norm_samples, density = True, color = 'lightblue', edgecolor = 'black', label = '$\overline{x} =$' + f' {np.mean(norm_samples):.3f}' + '\n$s =$' + f' {np.std(norm_samples):.3f}')
xmin, xmax = plt.xlim()
x = np.linspace(mean - 4*st_dev, mean + 4*st_dev, 10000)
p = stats.norm.pdf(x, mean, st_dev)
plt.plot(x, p, linewidth = 2, color = 'purple', label = 'Standard Normal\nDistribution')
plt.legend();
```
