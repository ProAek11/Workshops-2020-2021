# **Basic Operations**

## **Matrix Addition**

**Matrix addition is an operation which sums together matrices of the same dimension.**

$x + y =  
\begin{pmatrix}  
x_{11} & x_{12} & \cdots & x_{1 n} \\  
x_{21} & x_{22} & \cdots & x_{2 n} \\  
\vdots & \vdots & \ddots & \vdots \\  
x_{m1} & x_{m2} & \cdots & x_{mn}   
\end{pmatrix} + \begin{pmatrix}  
y_{11} & y_{12} & \cdots & y_{1 n} \\  
y_{21} & y_{22} & \cdots & y_{2 n} \\   
\vdots & \vdots & \ddots & \vdots \\  
y_{m1} & y_{m2} & \cdots & y_{mn}   
\end{pmatrix}$

<sup>Source: [Wikipedia](https://en.wikipedia.org/wiki/Matrix_addition)</sup>

```
arr_1 = np.array([[2,1,1],
                  [1,2,3]])

arr_2 = np.array([[0,2,3],
                  [4,0,0]])

arr_3 = np.array([[5,1,5],
                  [4,3,2]])
```

```
arr_1+arr_2
```

```
try:
  x + y
except:
  print('Can only add matrices if their dimensions are the same.')
```

### **Commutative Property**

$x + y = y + x$

```
arr_1+arr_2 == arr_2+arr_1
```

### **Associative Property**

$(x+y)+z = x+(y+z)$

```
(arr_1+arr_2)+arr_3 == arr_1+(arr_2+arr_3)
```

```
v1 = np.array([4,4])
v2 = np.array([-1,-2])
```

## **Dot Product (Matrix Multiplication)**

**The dot product or scalar product is an algebraic operation that takes two equal-length sequences of numbers (usually coordinate vectors), and returns a single number.**

$x \cdot y = \underset{i=1}{\overset{n}{\sum}}x_iy_i=x_1y_1 + x_2y_2 + ... + x_ny_n$

<sup>Source: [Wikipedia](https://en.wikipedia.org/wiki/Dot_product)</sup>

```
q = np.array([4,3])

r = np.array([4,1])

q.dot(r)
```

### **Commutative Property**

```
q.dot(r) == r.dot(q)
```

### **Associative Property (Does not hold for matrix multiplication)**

```
s = np.array([2,2])

q.dot(r.dot(s)) == s.dot(r.dot(q))
```

## **Finding the correlation between two data sets (Linear Regression)**

**Linear Regression - A statistical method that allows us to summarize and study relationships between quantitative variables: a dependent variable which is the feature we are trying to forecast and independent variable(s) which explain the dependent variable.**

<center>$y_i = \beta_0 + \beta_1x + \epsilon_i$<center/>

<sup>Reference: [Regression Methods from Penn State University](https://cython.readthedocs.io/en/latest/src/quickstart/overview.html)</sup>

```
study_array = np.array([[7, 2, 8, 8.5, 4, 9.25, 9.75, 7.5],
                       [8, 8, 7.75, 7.75, 6.75, 10, 11, 12],
                       [9.5, 10, 10.75, 8.75, 7.25, 7.75, 7, 2.5],
                       [9.25, 9, 9.25, 7.25, 7, 8, 8.5, 9.75],
                       [8, 4, 6.5, 8, 8.75, 9, 9, 9],
                       [7.25, 10, 8.75, 9.5, 7.75, 8, 9, 10.5],
                       [12, 12, 9.5, 2, 8.25, 10.5, 10.25, 7.5],
                       [11.75, 9, 8.75, 7.5, 12, 9, 8, 12.75],
                       [10, 7.25, 8.5, 8, 10, 6.75, 8.5, 9],
                       [5, 7.75, 9, 8, 8, 10.75, 10, 9],
                       [8, 8, 9, 8, 3.5, 4.75, 8.75, 9],
                       [8, 7.75, 6.75, 11, 10.25, 10, 10, 11],
                       [8.5, 8, 8.75, 12, 10, 7.5, 8.5, 10],
                       [8.75, 9.25, 7, 13, 2.75, 7.25, 10, 11.25],
                       [9, 8.25, 11, 8, 10, 10, 2.25, 9],
                       [9.5, 8.25, 10.25, 8.25, 11, 6.25, 12, 10.25],
                       [10.25, 7, 10, 9.5, 13, 8.75, 9.5, 13],
                       [7.25, 7.5, 9.75, 6.75, 12.25, 10.25, 6.75, 10],
                       [3, 3.25, 9.5, 9, 7.75, 7, 2.75, 1.5],
                       [8, 12.5, 9, 2.5, 8.5, 8, 8, 9.25],
                       [8, 8.75, 8, 10, 3.75, 3.75, 3.25, 8.25]])
```

```
lin_reg = stats.linregress(study_array.ravel(), one_hot.ravel())
print(lin_reg.intercept, lin_reg.slope)
```

```
plt.plot(study_array.ravel(), one_hot.ravel(), 'o', label='Data')
plt.plot(study_array.ravel(), lin_reg.intercept + lin_reg.slope*study_array.ravel(), 'r', label=f'Fitted Line\nR Squared: {lin_reg.rvalue:.2f}')
plt.yticks(np.arange(0,2,1))
plt.legend()
plt.xlabel('Study Hours')
plt.ylabel('Pass/Fail');
```

```
def pass_fail_forecast(x):
  '''Input the number of hours the student studied.'''
  result = lin_reg.intercept + lin_reg.slope * (x)
  
  if result >= .5:
    print('The model forecasts the student will pass.')
  else:
    print('The model forecasts the student will fail.')
  return result
```

```
pass_fail_forecast(5)
```

## **Video Games (Dot Product)**

**The dot product operation is used in video game and animation design. In video game design, the dot product is used to distinguish the relative position of different objects. For example, the dot product can tell the program that the player is standing in front of a door.**

**Example from Textbook:**

**For each of the following points x determine whether x is in front of or behind the NPC (Non-player character), where p = [-3,4] and v = [5,-2].**

- x = [0,0]
- x = [1,6]
- x = [-6,0]
- x = [-4,7]

<sup>Source: [3D Math Primer for Graphics and Game Development by Fletcher Dunn and Ian Parberry](https://www.taylorfrancis.com/books/mono/10.1201/b11152/3d-math-primer-graphics-game-development-fletcher-dunn-ian-parberry?refId=f1f4bd66-4434-4ade-8d68-24078f1e5f8c)</sup>

```
p = np.array([-3,4])
v = np.array([5,-2])

vector_list = [np.array([0,0]), np.array([1,6]), np.array([-6,0]), np.array([-4,7])]
```

```
for vector in vector_list:
  if v.dot(vector - p) > 0:
    print(f'The object x is in front of the NPC, dot product output {v.dot(vector - p)}')
  else:
    print(f'The object x is behind the NPC, dot product output {v.dot(vector - p)}')
```
